### 1 (a). importing the Important Libraries

```python
from bs4 import BeautifulSoup
         # By importing BeautifulSoup Soup we can parse and scrap the HTML and XML document
import requests
         # By Importing requests we can request to API and we can get the raw HTML contents from website i.e to scraping
import os
         # This library helps to intrect directly from Operating System
import nltk
         # By this we can import Natural Language toolkit
from nltk import corpus
         # By this we can import coprora reader which helps to handle specific corpus format 
import re
        # By this we can import Regular Expression which will help to find  some specific things from scrap
```

### 1(b). First taking a URL and store it in 'url1'

```python
url1='https://www.purpledocs.com/'
```

### 1(c). Making a request to access the URL

```python
html_content1=requests.get(url1).text
```

```python
html_content1
```

***here we get <Response [200]> that means we have succesfully accessed the URl***

```python
#html_content1=requests.get(url1).text
soup1=BeautifulSoup(html_content1,'html')  # here html is the type of document 
print(soup1.prettify())   #prettify() is used to  Beautiful Soup parse tree into a nicely formatted Unicode string
```

```python
for link in soup1.find_all('a'):   # Here we can find all the contents stored in directory 'a'
    print('Inner Text:{}'.format(link.text))  # Here all text of link is stored in '{}' and assigned as 'Inner Text' by using format command
    print('Title:{}'.format(link.get('title')))  # Here all title of link is stored in '{}' and assigned as 'Title'by using format command
    print('href:{}'.format(link.get('href')))  #  # Here all hypertexts  of link is stored in '{}' and assigned as 'href' by using format command
           # and printing all this we will get Output as:-
```

***Note: By Doing the same we can Extract the data of 'n' number of pages but here we have done only for 5 URL***

### 2.  Taking a second URL and store it in 'url2'

```python
url2='https://www.wimerasys.com/'
```

```python
html_content2=requests.get(url2).text
soup2=BeautifulSoup(html_content2,'lxml')
print(soup2.prettify())
```

```python
for link in soup2.find_all('a'):
    print('Inner Text:{}'.format(link.text))
    print('Title:{}'.format(link.get('title')))
    print('href:{}'.format(link.get('href')))
```
